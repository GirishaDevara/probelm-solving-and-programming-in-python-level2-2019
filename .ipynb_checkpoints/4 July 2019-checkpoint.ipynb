```python
# Linear graph

import matplotlib.pyplot as plt

plt.title("Basi linear plots")
plt.plot([1,2,3],[1,2,3])

plt.show()
```

```python
# Graph to plot distane and time

distance = list(range(2,11,2))
time = list(range(1,6))
# d = 2*t
plt.title("Distance Vs Time")
plt.plot(time, distance, 'b d')
plt.grid() # to get grid on graphical window
plt.xlabel("time")
plt.ylabel("Distance")
```

```python
# v = a t

t = list(range(1,11))
a = list(range(len(t)+1))
v = [a[i] * i for i in range(len(t))]
plt.plot(t,v,"g *")
plt.grid()
```

```python
import numpy as np
x = list(range(0,30))
a = np.sin(x)
plt.plot(x,a,'m-')
plt.grid()
```

```python
x = list(range(0,30))
a = np.sin(x)
b = np.cos(x)
plt.plot(a,b,'m-')
plt.grid()
```

```python
x = list(np.arange(0,4*np.pi,0.1))
# x = [i/10 for i in range(1,21)]
a = np.sin(x)
plt.plot(x,a,'g-d')
plt.grid()
```

```python
# plot the Income of Arkansasmfrom 2005 to 2013
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')

years = df.columns[2:]
incomeArkansas = df.values[3, 2:]

plt.title("Income of Arkansas")
plt.xlabel("year")
plt.ylabel("Income")
plt.plot(years,incomeArkansas)
plt.grid()
```

```python
# income of all states 2005 to 2013

avgIncome = df.iloc[:,2:].mean()
plt.plot(avgIncome,'g')
plt.title("All states Income")
plt.ylabel("Income")
plt.xlabel("Years")
plt.grid()
```

```python
def isprime(n):
    if n == 2:
        return True
    for i in range(2,int(n**0.5)+1):
        if n % i ==0:
            return False
    return True
        
small = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
primevalues=[]
for c in small:
    if isprime(ord(c)):
        primevalues.append(ord(c))
def magicword(s,n):
    result = ""
    for char in s[:n]:
        l = []
        for i in range(len(primevalues)):
            l.append((abs(primevalues[i]-ord(char)),primevalues[i]))
        #print(chr(min(l)[1]))
        result += chr(min(l)[1])
    print(result)
t = int(input())
for i in range(t):
    n = int(input())
    s = input()
    magicword(s,n)
```

```python
n = input()# 49
c = 0
for i in n:
    if i != "4" or i != "5":
        rem4 = int(i) % 4  
        rem5 = int(i) % 5
        if rem4 == "4" or rem4 == "5" or rem4 == 0:
            c += 1
        elif rem5 == "4" or rem5 == "5" or rem5 == 0:
            c += 1
    elif i == "4" or i =="5":
        c += 1
print(c)
        
```
